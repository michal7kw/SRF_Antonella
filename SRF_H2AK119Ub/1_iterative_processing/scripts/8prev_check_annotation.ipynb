```python
import pandas as pd

# Read the peak annotation file
peak_annotation = pd.read_csv('/beegfs/scratch/ric.broccoli/kubacki.michal/SRF_H2AK119Ub_cross_V5/SRF_H2AK119Ub/1_iterative_processing/analysis/8_annotation_and_enrichment/annotation_broad/tables/peak_annotation.csv')

```

```python
peak_annotation.head()
```

```python
# Get unique annotations and print them
unique_annotations = peak_annotation['annotation'].unique()

# Convert all intron and exon annotations to just "Intron" and "Exon"
unique_annotations_simplified = []
for annotation in unique_annotations:
    if annotation.startswith('Intron ('):
        if 'Intron' not in unique_annotations_simplified:
            unique_annotations_simplified.append('Intron')
    elif annotation.startswith('Exon ('):
        if 'Exon' not in unique_annotations_simplified:
            unique_annotations_simplified.append('Exon')
    else:
        unique_annotations_simplified.append(annotation)

print("Unique annotations found in the data:")
for annotation in sorted(unique_annotations_simplified):
    print(annotation)
```
